```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
df = pd.read_csv('../../results/ER_N100_d5_seed102_obj.csv')  # Adjust path if needed

# --- Boxplot: Distribution of Objective Values ---
plt.figure(figsize=(10, 6))
plt.boxplot([df['o1'], df['o2'], df['o3'], df['o4'], df['o5']], labels=['o1', 'o2', 'o3', 'o4', 'o5'])
plt.title('Distribution of Objective Values for Different Methods')
plt.xlabel('Method')
plt.ylabel('Objective Value')
plt.grid(True)
plt.show()

# --- Line Plot: Objective Values per Trial ---
plt.figure(figsize=(12, 6))
for idx in range(len(df)):
    plt.plot(['o1', 'o2', 'o3', 'o4', 'o5'], df.iloc[idx], alpha=0.2)
plt.title('Objective Values per Trial')
plt.xlabel('Method')
plt.ylabel('Objective Value')
plt.grid(True)
plt.show()
```

```python
# drop o3 column
df = df.drop(columns=['o3'])
```

```python
import numpy as np

# Set a style for better aesthetics
plt.style.use('ggplot')
plt.rcParams['font.sans-serif'] = ['SimHei']      # 用黑体显示中文
plt.rcParams['axes.unicode_minus'] = False        # 正常显示负号
plt.rcParams['mathtext.fontset'] = 'stix'

# --- Enhanced Boxplot ---
fig, ax = plt.subplots(figsize=(8, 6))

labels = ['$Random$', '$Post\\text{-}Opt$', '$Actual\\text{-}Policy$', '$Oracle\\text{-}Optimal$']
box = ax.boxplot(
    [df['o1'], df['o2'], df['o4'], df['o5']],
    labels=labels,
    patch_artist=True,
    boxprops=dict(facecolor='lightblue', color='blue'),
    medianprops=dict(color='red', linewidth=2),
    whiskerprops=dict(color='blue'),
    capprops=dict(color='blue')
)

# ax.set_title('不同目标函数值的分布情况', fontsize=16)
ax.set_xlabel('目标', fontsize=14)
ax.set_ylabel('目标函数值', fontsize=14)
ax.grid(True, linestyle='--', alpha=0.7)

plt.tight_layout()
plt.savefig("fig/obj_test100_boxplot.png", dpi=300, bbox_inches='tight')
plt.show()

# --- Enhanced Line Plot ---
fig, ax = plt.subplots(figsize=(10, 6))

colors = plt.cm.viridis(np.linspace(0, 1, len(df)))

for idx in range(len(df)):
    ax.plot(labels, df.iloc[idx], alpha=0.15, color='grey')

# Overlay the average trend
mean_values = df.mean()
ax.plot(labels, mean_values, color='black', linewidth=2.5, label='平均趋势')

# ax.set_title('每次实验的目标函数值变化', fontsize=16)
ax.set_xlabel('目标', fontsize=14)
ax.set_ylabel('目标函数值', fontsize=14)
ax.legend()
ax.grid(True, linestyle='--', alpha=0.7)

plt.tight_layout()
plt.savefig("fig/obj_test100_lineplot.png", dpi=300, bbox_inches='tight')
plt.show()
```
